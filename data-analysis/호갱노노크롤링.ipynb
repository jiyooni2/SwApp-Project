```python
import csv
addr_list = []
apt_list = []
apt_nm_list = []

with open("C:\\Users\\kjy77\\Downloads\\apart_trade20_10_21_09.csv", 'r', encoding='CP949') as data_file2:
    data_lines = csv.reader(data_file2, delimiter=',')
    for idx, data_line in enumerate(data_lines):
        if idx > 17:
            trade_area = data_line[5]
            if (99 <= float(trade_area)) and (float(trade_area) <= 132):
                #30평대만 추출
                pass
            else:
                continue
            tmp = data_line[0].split()
            doro_addr = tmp[0]+" "+tmp[1]
            doro_addr += " "+data_line[11]
            trade_date = data_line[6]
            trade_price = data_line[8]
            trade_floor = data_line[9]
            year = data_line[10]
            trade_year = trade_date[0:4]
            trade_month = trade_date[4:6]
            apt_nm = data_line[4]
            addr_dong = data_line[0].split()[2]
            addr_gu = data_line[0].split()[1]
            trade_price = trade_price.replace(",", "")

            dong_apt_nm = addr_dong+" "+apt_nm
            addr = data_line[0]+" "+data_line[1]

            if len(apt_nm_list) == 0:
                apt_nm_list.append([dong_apt_nm, year, addr])
            else:
                if apt_nm_list[len(apt_nm_list)-1] != [dong_apt_nm, year, addr]:
                    apt_nm_list.append([dong_apt_nm, year, addr])

            if len(addr_list) == 0:
                addr_list.append(doro_addr)
            else:
                if addr_list[len(addr_list)-1] != doro_addr:
                    addr_list.append(doro_addr)
#             apt_list밑에랑 중복변수임 중복변수 더 있는지 찾아보기
            apt_list.append(
                [doro_addr, trade_year, trade_month, trade_area, trade_floor, trade_price])
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import re


chromedriver="C:\\Users\\kjy77\\1_Python\\chromedriver.exe"

# options = webdriver.ChromeOptions()
# options.add_argument('headless')
# options.add_argument('window-size=1920x1080')
# options.add_argument("disable-gpu")
# options.add_argument("lang=ko_KR")

driver=webdriver.Chrome(chromedriver)
#driver=webdriver.Chrome(chromedriver,options=options)


driver.get("http://www.hogangnono.com")
#driver.quit()
```

```python
import csv

data_file = open("apt_list.csv", 'w', encoding='utf-8-sig', newline='')
data_write = csv.writer(data_file, delimiter=',')
# data_write.writerow(['아파트', '241세대', '2009년 3월(13년차)', '6개동', '최고 15층',
#                     '용적률 209%', '건폐율 23%', '지하주차 286대', '개별난방', '도시가스', '내진설계대상', '평지에 있는 단지'])
# data_file.close()

#넣을때는 그냥 이렇게 넣고 sql에 집어넣을때 제한조건을 걸어주자
```

```python
count = 0
not_list = []
crawling_error_list = []
completed_apt_list = []
for idx, apt in enumerate(apt_nm_list):
    is_find = False
    try:
        elem = driver.find_element_by_css_selector(
            'div.keyword-group div input.keyword')
        elem.clear()
        org_apt_name = apt[0]
        apt_name = apt[0]
        apt_addr = apt[2]
        elem.send_keys(apt_name)
        # 여기서 동+아파트이름으로 넘겨주기
        elem.send_keys(Keys.RETURN)
        time.sleep(3)

        year = apt[1]

        apt_list = driver.find_elements_by_css_selector('li.apt')
        # 검색 후 검색결과창에서, 아파트 선택
        for apt in apt_list:
            try:
                hogang_year = apt.find_element_by_css_selector(
                    'div.desc span.startDate').text
                hogang_year = hogang_year.split("년")[0]

                selected_apt = apt.find_element_by_css_selector('a')
                selected_apt_name = apt.find_element_by_css_selector(
                    'div span').text
               # 검색결과창에서 뜨는 세대수
                selected_dong = selected_apt_name.split()[0]
                if (selected_dong == apt_name.split()[0]) and (int(hogang_year) == int(year)):
                    # 동이름과 연도일치
                    apt_href = selected_apt.get_attribute('href')
                    driver.get(apt_href)
                    # time.sleep(3)
                    hogang_addr = driver.find_element_by_css_selector(
                        'div.header-info h2.address').text
                    # 주소까지 일치하면 맞는 case
                    if hogang_addr == apt_addr:
                        is_find = True
                        try:
                            apt_info = [org_apt_name, selected_apt_name]
                            base_infos = driver.find_elements_by_css_selector(
                                "div.infoCard div.baseInfo ul li")
                            action = ActionChains(driver)
                            action.move_to_element(
                                base_infos[len(base_infos)-1]).perform()

                            for base_info in base_infos:
                                apt_info.append(base_info.text)

                            completed_apt_list.append(apt_info)
                            # completed_apt_list에 결과물 들어있음(csv랑 같음)
                            data_write.writerow(apt_info)
                            print(idx+1, "/", len(apt_nm_list),
                                  selected_apt_name, "SUCCESS")
                        except Exception as e:
                            print(e, "기본 Info 뽑기 실패")
                            crawling_error_list.append(
                                [apt_name, year, apt_addr])

                        break
                    else:
                        # 동이름과 연도 일치하지만 주소 다름
                        print(hogang_addr, apt_addr)
                        break
            except Exception as e:
                pass
#                 print("연도/주소 크롤링 error", e)
    except Exception as e:
        print("검색과정 error", e)

    if is_find == False:
        print(count, "/", idx, apt_name, "Failed")
        count += 1
        not_list.append([apt_name, year, apt_addr])
        continue

#     driver.get(apt_href)
#     time.sleep(3)
```
