```python
def find_location(region, page_num):
	url = 'https://dapi.kakao.com/v2/local/search/keyword.json'
	params = {'query': region, 'page' : page_num}
	headers = {"Authorization": "KakaoAK d000ae5578cd1ffb3a231b3aa82162e9"}

	places = requests.get(url, params=params, headers=headers).json()['documents']
	return places
```

```python
def find_info(places):
	coordinate = []
	try:
		place = places[0]
		coordinate.append(place['place_name'])
		coordinate.append(float(place['y']))
		coordinate.append(float(place['x']))
	except:
		return coordinate
	return coordinate
```

```python
def keywords(location):
	local_name = find_location(location, 1)
	coordinate = find_info(local_name)
	return coordinate
```
