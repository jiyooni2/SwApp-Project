```python
#apt 테이블 생성
sql="""
CREATE TABLE apt (
apt_code INT UNSIGNED NOT NULL,
x_code DECIMAL(13, 10),
y_code DECIMAL(13, 10),
apt_nm VARCHAR(50) NOT NULL,
co_wo VARCHAR(40),
heat_mthd VARCHAR(10),
dorojuso VARCHAR(50),
use_confm_de VARCHAR(30),
all_hshld_co INT UNSIGNED,
adres_gu VARCHAR(10),
adres_dong VARCHAR(10),
year INT UNSIGNED,
parking_space VARCHAR(5),
near_price INT UNSIGNED,
PRIMARY KEY(apt_code)
);
"""
print(sql)
cursor.execute(sql)
db.commit()
```

```python
#price table 생성
sql="""
CREATE TABLE price (
no INT AUTO_INCREMENT NOT NULL PRIMARY KEY,
apt_code INT UNSIGNED NOT NULL,
date DATE NOT NULL,
price INT UNSIGNED NOT NULL,
maintenance_cost INT UNSIGNED,
transfer_tax INT UNSIGNED,
tax INT UNSIGNED,
pyeong INT UNSIGNED NOT NULL,
FOREIGN KEY(apt_code) references apt(apt_code)
);
"""
    
print(sql)
cursor.execute(sql)
db.commit()
```

```python
# 세대수랑 동이랑 아파트이름
# 세대수 없는거
# 신내역금강펜테리움센트럴파크
# 해링턴스퀘어
# 녹번역e편한세상캐슬
# sql query = select * from apt where all_hshld_co is null;

print("이름", "연도", "주차공간", "근처시세")

for data in sql_result:
    apt_name = data[2]+data[0]
    apt_name = apt_name.replace(" ", "")

    elem = driver.find_element_by_css_selector(
        'div.keyword-group div input.keyword')
    elem.clear()
    elem.send_keys(apt_name)
    # 여기서 동+아파트이름으로 넘겨주기
    elem.send_keys(Keys.RETURN)
    time.sleep(2)

    households = data[1]

    apt_list = driver.find_elements_by_css_selector('li.apt')
    # 검색 후 검색결과창에서, 아파트 선택

    for apt in apt_list:
        is_find = False
        try:
            hogang_households = apt.find_element_by_css_selector(
                'div span.household').text.replace("세대", "").replace(",", "")
           # 검색결과창에서 뜨는 세대수
            if int(hogang_households) == households:
                selected_apt = apt.find_element_by_css_selector('a')
                apt_href = selected_apt.get_attribute('href')
                is_find = True
                break
        except:
            pass

    if is_find == False:
        print(apt_name, households, "일치하는 아파트 찾을 수 없음, 세대수, 아파트 다름")
        continue

        
    
    driver.get(apt_href)
    time.sleep(3)
    
    try:
        year = driver.find_elements_by_css_selector(
            'div.card>div.css-e81zx8>ul>li')[1]
        action = ActionChains(driver)
        action.move_to_element(year).perform()
        year = year.text
        year = year.split("년")[0]
    except:
        year="NULL"
        
    if "%" in year:
        year="NULL"

    
    try:
        parking_space = get_result('div.parking>span>span.status')
    except:
        parking_space = "NULL"


    try:
        near_price = get_result('div.region-info>span.price')

        near_price = near_price.replace("억", "")
        if float(near_price) % 1 == 0:
            near_price = near_price+"0000"
            near_price = near_price.replace(".", "")
        else:
            near_price = near_price+"000"
            near_price = near_price.replace(".", "")
    except:
        near_price = str(0)
    #data[3] : apt_code

    sql = "UPDATE apt SET year="+year+", parking_space='"+parking_space + \
        "', near_price='"+near_price+"'"+" WHERE apt_code="+str(data[3])+";"
    cursor.execute(sql)
    
    db.commit()
    
    print(apt_name, year, parking_space, near_price)
```
