```python
import csv
addr_list = []
apt_list = []
apt_nm_list = []

with open("./apart_trade20_10_21_09.csv", 'r', encoding='CP949') as data_file2:
    data_lines = csv.reader(data_file2, delimiter=',')
    for idx, data_line in enumerate(data_lines):
        if idx > 1:
            trade_area = data_line[5]
            if (99 <= float(trade_area)) and (float(trade_area) <= 132):
                #30평대만 추출
                pass
            else:
                continue
            tmp = data_line[0].split()
            doro_addr = tmp[0]+" "+tmp[1]
            doro_addr += " "+data_line[11]
            trade_date = data_line[6]
            trade_price = data_line[8]
            trade_floor = data_line[9]
            year = data_line[10]
            trade_year = trade_date[0:4]
            trade_month = trade_date[4:6]
            apt_nm = data_line[4]
            addr_dong = data_line[0].split()[2]
            addr_gu = data_line[0].split()[1]
            trade_price = trade_price.replace(",", "")

            dong_apt_nm = addr_dong+" "+apt_nm
            addr=data_line[0]+" "+data_line[11]
            
            if len(apt_nm_list) == 0:
                apt_nm_list.append([dong_apt_nm, year, addr])
            else:
                if apt_nm_list[len(apt_nm_list)-1] != [dong_apt_nm, year, addr]:
                    apt_nm_list.append([dong_apt_nm, year, addr])

            if len(addr_list) == 0:
                addr_list.append(doro_addr)
            else:
                if addr_list[len(addr_list)-1] != doro_addr:
                    addr_list.append(doro_addr)
#             apt_list밑에랑 중복변수임 중복변수 더 있는지 찾아보기
            apt_list.append(
                [doro_addr, trade_year, trade_month, trade_area, trade_floor, trade_price])
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver import ActionChains
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
import re


chromedriver="C:\\Users\\kjy77\\1_Python\\chromedriver.exe"

# options = webdriver.ChromeOptions()
# options.add_argument('headless')
# options.add_argument('window-size=1920x1080')
# options.add_argument("disable-gpu")
# options.add_argument("lang=ko_KR")

driver=webdriver.Chrome(chromedriver)
#driver=webdriver.Chrome(chromedriver,options=options)


driver.get("https://kbland.kr/map")
#driver.quit()
```

```python
apt_price_list = []
success_list = []
failed_list = []

for index, apt in enumerate(apt_nm_list):
    try:
        address = apt[2]
        apt_name = apt[0]

        search_button = driver.find_element_by_css_selector(
            'button.btn.btn-search').click()
        search_query = driver.find_element_by_css_selector(
            'div.inp-total-search>input')
        search_query.clear()
        search_query.send_keys(address)  # address 입력
        search_query.send_keys(Keys.RETURN)

        time.sleep(2)
        try:
            search_list = driver.find_elements_by_css_selector(
                "div#tabHscm div.item-search-poi")
            search_list[0].find_element_by_css_selector(
                "div>button.btn.search-poi").click()
        except:
            pass
        time.sleep(2)

        name_insite = driver.find_element_by_css_selector(
            "div.summaryContW>strong.tit>span.txt2").text
        driver.find_element_by_css_selector(
            "div.summaryContW>strong.tit>span.txt2").click()
#         driver.find_element_by_css_selector('span.sBot').click()

        # 평형 바꾸기
        time.sleep(2)
        is_find = False
        pyeong = driver.find_element_by_css_selector("div.select-text").text
        pyeong = float(pyeong.split("/")[1].strip().replace("m²", ""))
        pyeong = pyeong/3.3
        if 30 <= pyeong and pyeong < 40:
            is_find = True
            pass
        else:
            more_button = driver.find_element_by_css_selector(
                "button.btn.btn-select").click()
            time.sleep(2)
            pyeong_list = driver.find_elements_by_css_selector(
                "div.ta-body > div.ta-tr")
            for p in pyeong_list:
                p_all = p.find_elements_by_css_selector("div.ta-cell")

                p2 = p_all[0].find_element_by_css_selector(
                    "div.inner div.s-row span.txt-l")
                p2_text = int(p2.text.split("/")[1].strip().split(".")[0])
                p2_text = p2_text/3.3

                p2_price = p_all[1].find_element_by_css_selector(
                    "div.inner div.s-row span").text
                if (30 <= p2_text and p2_text < 40 and p2_price != "0"):
                    p2.click()
                    is_find = True
                    break
        pyeong = int(p2_text)

        if(is_find == False):
            raise Exception("pyeong not matching")

        title_price = driver.find_element_by_css_selector(
            "div.kb-price strong.tit-cost span").text

        if(title_price == "조사 대상 아님" or title_price == "시세없음"):
            raise Exception("No price data")

        time.sleep(2)
        # 시세보기 버튼 클릭
        price_button = driver.find_element_by_css_selector(
            'button.btn.btn-barbtn')
        action = ActionChains(driver)
        action.move_to_element(price_button).click(price_button).perform()

        time.sleep(3)

        years = driver.find_elements_by_css_selector(
            'div.priceyear>div.p-yeracheck label.btn')

        for idx, year in enumerate(years):
            if idx == 0:
                continue
            if idx >= 4:
                break
            year.click()

        price_table = driver.find_elements_by_css_selector("div.data-table")

        if(len(price_table) == 2):
            price_table = price_table[1]
        else:
            price_table = price_table[0]

        per_year_table = price_table.find_elements_by_css_selector(
            "table>tbody>tr>td")

        for idx, table in enumerate(per_year_table):
            year = table.find_element_by_css_selector(
                "div.data-wrap>div.table-label").text
            prices = table.find_elements_by_css_selector(
                "div.table-data.average > div.table-line")

            for price in prices:
                month = price.find_element_by_css_selector("div.data-month")
                avgs = price.find_elements_by_css_selector("div.data-average")
                # avgs[0] : 하위평균 [1] : 평균 [2] : 상위평균

                tmp = [name_insite, apt_name, pyeong, year, month.text,
                       avgs[0].text, avgs[1].text, avgs[2].text]
                apt_price_list.append(tmp)

        inner_close_button = driver.find_element_by_css_selector(
            "button.btn.btn-pageback").click()
        time.sleep(1)
        close_button = driver.find_element_by_css_selector(
            'button.btn.btn-close').click()  # 다음 검색 위해서 close
        print(index+1, "/", len(apt_nm_list), apt_name, "success")
        success_list.append(apt_name)
    except Exception as err:
        failed_list.append(apt_name)
        driver.get("https://kbland.kr/map")
        print(apt_name, "failed", err)
    finally:
        time.sleep(3)


# Code 최적화
# 1. 도로명 주소 쓰기
# 2. 시세 없음이라 뜨는것 필터링해서 바로 conitnue 진행
```
